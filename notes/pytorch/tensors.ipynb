# Basic Tensor Concepts

Dimensions & Shapes

```python
scalar = torch.tensor(5)

vector = torch.tensor([1, 2, 3])

matrix = torch.tensor([[1., 2., 3.], 
                       [4., 5., 6.]])

tensor_3d = torch.tensor([[[1, 2], [3, 4]],
                          [[5, 6], [7, 8]]])
                          
```

```python
print(f"Scalar - Shape: {scalar.shape}, Rank: {len(scalar.shape)}")
print(f"Vector - Shape: {vector.shape}, Rank: {len(vector.shape)}")
print(f"Matrix - Shape: {matrix.shape}, Rank: {len(matrix.shape)}")
print(f"3D Tensor - Shape: {tensor_3d.shape}, Rank: {len(tensor_3d.shape)}\n")
```

# Matrix Operations

```python
# Create two matrices for multiplication
A = torch.tensor([[1, 2, 3], 
                  [3, 4, 5]])
B = torch.tensor([[5, 6, 7], 
                  [7, 8, 9]]).transpose(0, 1)

# Demonstrate matrix multiplication
C = torch.matmul(A, B)  # or A @ B

print("Matrix Multiplication (m x n @ n x p = m x p):")
print(f"Shape A: {A.shape}")
print(f"Shape B: {B.shape}")
print(f"Shape Result: {C.shape}\n")

print("Matrix B after transposition:")
print(B)
```

# Dimensions

```python
# Create a tensor for dimension operations
data = torch.randn(4, 3, 2)

# Demonstrate different dimension operations
print("Original tensor shape:", data.shape)
print("Mean along dim=0 shape:", data.mean(dim=0).shape)
print("Mean along dim=-1 shape:", data.mean(dim=-1).shape)

# Visualization of dimension operations
plt.figure(figsize=(15, 5))

# Original matrix
plt.subplot(131)
sns.heatmap(matrix, annot=True, fmt='.2f', cmap='viridis')
plt.title('Original Matrix')

# Mean along dim=0
plt.subplot(132)
sns.heatmap(matrix.mean(dim=0).reshape(1, -1), annot=True, fmt='.2f', cmap='viridis')
plt.title('Mean along dim=0')

# Mean along dim=1
plt.subplot(133)
sns.heatmap(matrix.mean(dim=1).reshape(-1, 1), annot=True, fmt='.2f', cmap='viridis')
plt.title('Mean along dim=1')
plt.tight_layout()
plt.show()
```
