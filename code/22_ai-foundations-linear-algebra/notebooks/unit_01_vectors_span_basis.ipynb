## Standard Basis vs Changing Basis

The standard basis for R^2 is `e1 = (1, 0)` and `e2 = (0, 1)`. When you write a vector like `(5, 7)`, you are usually giving its coordinates in that standard basis:

```text
(5, 7) = 5 * e1 + 7 * e2
       = 5 * (1, 0) + 7 * (0, 1)
```

Changing basis means describing the same geometric vector using a different set of basis vectors. If your new basis is `v1, v2, ..., vn`, you are looking for coefficients `c1, c2, ..., cn` such that:

```text
target = c1 * v1 + c2 * v2 + ... + cn * vn
```

In matrix form, put the basis vectors into the columns of a matrix `A`:

```text
A = [v1 v2 ... vn]
A @ c = target
```

Solving `A @ c = target` gives you the coordinates of `target` in the new basis. In PyTorch, that is:

```python
c = torch.linalg.solve(A, target)
reconstruction = A @ c
```

This is the core math behind the Manim basis transformation video: the vector itself can stay fixed in space while the coordinate description changes because the basis vectors changed. Visually, the grid changes; algebraically, the coordinate vector `c` changes.

## Things to Notice

- Order of basis vectors matters for the column order of `A`. If you stack `v1, v2, v3` as columns in that order, the solver returns `(c1, c2, c3)` in that order. Swap columns, swap coefficient order.
- Floats vs ints matter. `torch.linalg.solve` wants floats. If you build a tensor from integer literals like `torch.tensor([[1, 2], [1, 3]])`, it will be an integer tensor and the solver complains. Use `1.` instead of `1`, or cast with `.float()`.
- Reconstruction is your safety net. Always do `A @ c` and check it equals the target. If it doesn't, you've made an error somewhere, most likely transposing rows vs columns when building `A`.
